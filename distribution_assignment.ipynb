Q.3  Write a Python function to calculate the probability density function of a normal distribution with 
given mean and standard deviation at a given point.

```python
import math
import matplotlib.pyplot as plt
import seaborn as sns

def normal_pdf(x,mean,std):
    coeff=1/(std * math.sqrt(2*math.pi))
    #print('coefficiant:',coeff)
    expo=-((x-mean)**2)/(2*std**2)
    pdf=coeff*math.exp(expo)
    return pdf
```

Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 
and plot a histogram of the results using matplotlib.

```python
import numpy as np
```

```python
#params of binomial distribution
n=10  #no of trials
p=0.4  #prob of succss on each trial
```

```python
#generate random samples of size 1000 from binomial distribution
size=1000
random_data=np.random.binomial(n,p,size)
#print(random_data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(random_data,bins=10)
plt.xlabel('No of Success')
plt.ylabel('Frequency')

plt.show()

sns.histplot(random_data,kde=True)
```

6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution 
with given mean at a given point.

```python
import math
```

```python
#exam_score=[75,83,92,64,78,85,90,71,79,87,68,95,81,89,76,84,88,72,77,93]
```

```python
def cdf_poisson(lamb,point):
    cdf=0
    for i in range(point+1):
        
        cdf+=(math.exp(-lamb) * (lamb ** i))/math.factorial(i)
        #print(cdf)
    
    return cdf
```

```python
lamb=3
point=5

cdf=cdf_poisson(lamb,point)
print(cdf)
```

```python
from scipy.stats import poisson
```

```python
cdf1=poisson.cdf(lamb,point)
print(cdf1)
```
